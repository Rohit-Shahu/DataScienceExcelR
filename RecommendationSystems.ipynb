```python
books = books.drop_duplicates(subset=['UserID','BookTitle'])
```

```python
user_books_df = books.pivot(index='BookTitle',
                                 columns='UserID',
                                 values='BookRating')
```

```python
#Impute those NaNs with 0 values
user_books_df.fillna(0, inplace=True)
```

```python
cos_sim = cosine_similarity(user_books_df)
```

```python
cos_sim
```

Since every book is similar to itself we can replace the values across the diagonal to 0

```python
np.fill_diagonal(cos_sim,0)
```

```python
cos_sim
```

Now that we have the cosine similarity matrix with us we can now make use of this and start recommending similar books

```python
def recommend_similar_books(book_name):
    #retrieving the index of the book name given as parameter
    index = np.where(user_books_df.index == book_name)[0][0]
    #retrieving all the values from cos_sim of the index
    distance = cos_sim[index]
    '''
    retrieving first 5 sorted values(descending - to get the books with nearest distance values with the book)
    '''
    similar_items = sorted(list(enumerate(cos_sim[index])),key=lambda x:x[1],reverse=True)[0:5]
    for i in similar_items:
        print(user_books_df.index[i[0]])
```

```python
recommend_similar_books('iI Paradiso Degli Orchi')
```

```python
recommend_similar_books('stardust')
```

```python
recommend_similar_books("McNally's Chance")
```
